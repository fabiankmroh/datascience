## `isin()`

```python
cond = train["Embarked"].isin(["Q", "S"])

# Extract Data that has "Q" & "S" for "Embarked" category
train[cond].head()
```

## `.str.contains()`

```python
str_contains = train["Ticket"].str.contains("STON")

train[str_contains].head()
```

```python
# See only the designated column using `loc`

train.loc[str_contains, ["Ticket"]].head()
```

## Null

```python
null_age = train["Age"].isnull()

train.loc[null_age, ["Age"]].head() 
```

`NaN`: Not A Number

```python
null_age = train["Age"].notnull()

train.loc[null_age, ["Age"]].head()
```
